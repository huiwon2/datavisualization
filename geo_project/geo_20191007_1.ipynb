```python
#log변환 - 완료
#소형
#print(seoulapt['소형 거래금액(평균)'])
seoulapt['소형 거래금액(평균)'].apply(lambda x: float(x))
seoulapt['소형 거래금액(평균)'] = np.log(seoulapt['소형 거래금액(평균)'])

 #중소형
seoulapt['중소형 거래금액(평균)'].apply(lambda x: float(x))
seoulapt['중소형 거래금액(평균)'] = np.log(seoulapt['중소형 거래금액(평균)'])
 #중형
seoulapt['중형 거래금액(평균)'].apply(lambda x: float(x))
seoulapt['중형 거래금액(평균)'] = np.log(seoulapt['중형 거래금액(평균)'])
 #중대형
seoulapt['중대형 거래금액(평균)'].apply(lambda x: float(x))
seoulapt['중대형 거래금액(평균)'] = np.log(seoulapt['중대형 거래금액(평균)'])
 #대형
seoulapt['대형 거래금액(평균)'].apply(lambda x:float(x))
seoulapt['대형 거래금액(평균'] = np.log(seoulapt['대형 거래금액(평균)'])
```

```python
seoulapt.head()
```

```python
#강동구 내(둔촌제2동, 둔촌제1동, 길동, 천호제3동, 천호제2동, 천호제1동) 연도 별(2020-2018) 거래금액(평균) 비교하기
#거래금액 연도 별 추이 비교시 소형, 중형, 중소형, 대형 별 금액 추이?
#가로 막대 그래프로 표시하기
find_gangdong = seoulapt_20to18.loc[(seoulapt['자치구 코드'] == 740)]
print(find_gangdong)
```

```python
#find_gangdong(강동구의 20년-18년 정보의 개수를 54개로 뽑음)
print(find_gangdong.info())
```

```python
#log변환 - 완료
#소형
#print(seoulapt['소형 거래금액(평균)'])
find_gangdong['소형 거래금액(평균)'] = np.log(find_gangdong['소형 거래금액(평균)'])

#중소형
find_gangdong['중소형 거래금액(평균)'] = np.log(find_gangdong['중소형 거래금액(평균)'])
#중형
find_gangdong['중형 거래금액(평균)'] = np.log(find_gangdong['중형 거래금액(평균)'])
#중대형
find_gangdong['중대형 거래금액(평균)'] = np.log(find_gangdong['중대형 거래금액(평균)'])
#대형
find_gangdong['대형 거래금액(평균'] = np.log(find_gangdong['대형 거래금액(평균)'])
```

```python
find_gangdong.head()
```

```python
#선그래프
#소형 거래금액 부분
#둔촌제2동
#동별로 나눠서 그래프 그리기(둔촌제2동, 둔촌제1동, 길동, 천호동)
df_gangdong = find_gangdong[['소형 거래금액(평균)', '거래년도']]
df_dunchon2 = df_gangdong[(find_gangdong['행정동 명'] == '둔촌제2동')]
# 소형 거래금액(평균)를 x축으로, 거래년도를 y축으로 하는 선 그래프
plt.plot(df_dunchon2['거래년도'], df_dunchon2['소형 거래금액(평균)'])

# 그래프 제목과 축 라벨 설정
plt.title('둔촌제2동')
plt.xlabel('거래년도')
plt.ylabel('소형 거래금액(평균)')


# 그래프 보여주기
plt.show()
#그래프 결과=>2018년부터 금액이 우상향 하고 있음을 나타냄
```
